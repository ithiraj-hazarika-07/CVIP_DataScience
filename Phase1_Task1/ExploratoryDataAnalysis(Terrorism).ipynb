**Importing python libraries for data analysis**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

**Importing and loading the Terrorism Dataset obtained from Kaggle**

```python
# Dataset Link: https://drive.google.com/file/d/15ILqVulIRA2BNEkUwXdRB-o_aH4Vfkzw/view?usp=sharing

from google.colab import drive
drive.mount('/content/drive')
```

```python
url = '/content/drive/MyDrive/terrorismData.csv'
terrorism = pd.read_csv(url)
```

```python
terrorism
```

**Getting details of the Terrorism Dataset**

```python
terrorism.info()
```

**Renaming columns with appropriate names for futher analysis**

```python
terrorism = terrorism.rename(columns={'Target_type': 'TargetType', 'Weapon_type': 'WeaponType'})
terrorism.tail(3)
```

**Checking for null values (if any) in the dataset**

```python
missingValue = terrorism.isna().sum()
print(missingValue)
```

**Changing 'NaN' values of Killed and Wounded columns to '0' and converting datatype from 'float64' to 'int64'**

```python
terrorism.Killed = terrorism.Killed.fillna(0)
terrorism.Wounded = terrorism.Wounded.fillna(0)

terrorism.Killed =  terrorism.Killed.astype(int)
terrorism.Wounded =  terrorism.Wounded.astype(int)

terrorism.info()
```

**Replacing the 'NaN' with 'Unknown'**

```python
terrorism = terrorism.fillna('Unknown')
```

```python
missingValue = terrorism.isna().sum()
print(missingValue)
```

### **Countries which experienced Terror Attacks till 2017**

```python
i = terrorism.shape[0]

terror_countries = set()

for row in range(i):
  terror_countries.add(terrorism.Country[row])

print('List of countries that experienced terror attacks :')
for country in sorted(terror_countries):
  print(country)

print('\nTotal Number of countries that experienced terror attacks :',len(terror_countries))
```

**Visualization of the Countries against no. of Terror Attacks**

```python
print('Country with most terror attacks :', terrorism.Country.value_counts().idxmax())
print('Country with least terror attacks :', terrorism.Country.value_counts().idxmin(),'\n')

countryName = terrorism.Country.value_counts()[:20]
print(countryName, '\n')
plt.figure(figsize=(14, 6))
sb.barplot(x = countryName.values, y = countryName.index)
plt.ylabel('<-- Countries -->')
plt.xlabel('Number of Terror Attacks -->')
plt.title('Top 20 Countries affected by Terror Attacks')
plt.grid(True)
plt.show()
```

**Checking for Cities that experienced most Terror Attacks**



```python
print('City with most terror attacks :', terrorism.City.value_counts().index[1],'\n')

city = terrorism.City.value_counts()[:20]
print(city, '\n')
plt.figure(figsize=(14, 6))
sb.barplot(x = city.values, y = city.index)
plt.ylabel('<-- City -->')
plt.xlabel('Number of Terror Attacks -->')
plt.title('Top 20 cities affected by Terror Attacks')
plt.grid(True)
plt.show()
```

**Checking for States that experienced most Terror Attacks**

```python
print('State with most terror attacks :', terrorism.State.value_counts().idxmax())
print('State with most terror attacks :', terrorism.State.value_counts().idxmin(),'\n')

state = terrorism.State.value_counts()[:20]
print(state, '\n')
plt.figure(figsize=(14, 6))
sb.barplot(x = state.values, y = state.index)
plt.ylabel('<-- State -->')
plt.xlabel('Number of Terror Attacks -->')
plt.title('Top 20 States affected by Terror Attacks')
plt.grid(True)
plt.show()
```

### **Year-wise distribution of Terrorist Activities**

```python
print('Year with highest no. of Terror Attacks :', terrorism.Year.value_counts().idxmax())
print('Year with lowest no. of Terror Attacks :', terrorism.Year.value_counts().idxmin(),'\n')

x_year = terrorism.Year.unique()
y_count_years = terrorism.Year.value_counts().sort_index()
plt.figure(figsize=(16, 6))
sb.barplot(x=x_year, y=y_count_years)
plt.xticks(rotation=60)
plt.xlabel('Attack Year')
plt.ylabel('Number of Attacks each year -->')
plt.title('Terror attacks between 1970 and 2017')
plt.show()
```

### **Total no. of people killed till 2017**

```python
killed = terrorism.loc[:, 'Killed']
print('Total killed till 2017 :', sum(killed))
```

```python
attack = terrorism.loc[:, 'AttackType']
killedData = pd.concat([attack, killed], axis = 1)

newkilledData = killedData.pivot_table(columns='AttackType', values='Killed', aggfunc='sum')
newkilledData
```

### **Total no. of people wounded till 2017**

```python
wounded = terrorism.loc[:, 'Wounded']
print('Total wounded till 2017 :', sum(wounded))
```

```python
attack = terrorism.loc[:, 'AttackType']
woundedData = pd.concat([attack, wounded], axis = 1)

newwoundedData = woundedData.pivot_table(columns='AttackType', values='Wounded', aggfunc='sum')
newwoundedData
```

### **Obtain the details of the Terror Attacks performed by a Terrorist Group**

```python
find = input('Enter a terrorist group name : ')
groupDetails = terrorism[terrorism['Group'] == find]
groupDetails.head()
```

```python
print('Year with highest no. of Terror Attacks :', groupDetails.Year.value_counts().idxmax())
print('Year with lowest no. of Terror Attacks :', groupDetails.Year.value_counts().idxmin(), '\n')

killed = groupDetails.loc[:, 'Killed']
print('Total killed till 2017 :', sum(killed))
wounded = groupDetails.loc[:, 'Wounded']
print('Total wounded till 2017 :', sum(wounded), '\n')


x_year = groupDetails.Year.unique()
y_count_years = groupDetails.Year.value_counts().sort_index()
plt.figure(figsize=(10, 6))
sb.barplot(x=x_year, y=y_count_years)
plt.xticks(rotation=60)
plt.xlabel('Attack Year')
plt.ylabel('Number of Terror Attacks each year -->')
plt.title(f'Terror attacks between 1970 and 2017 caused by {find}')
plt.show()
```
